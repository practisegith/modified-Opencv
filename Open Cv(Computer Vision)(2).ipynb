```python
import numpy as np
import cv2

from matplotlib import pyplot as plt
```

```python
#Loading Image

image = cv2.imread('flowerforcv.jpg')
```

```python
#Display Image

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
           
plt.title("original image")        
```

```python
#Converting Colour image to Gray

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(gray,cv2.COLOR_BGR2RGB))

plt.title("gray image")
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

plt.imshow(cv2.cvtColor(blurred_image,cv2.COLOR_BGR2RGB))

plt.title("blurred image")
```

```python
# Detect edges in the image using Canny edge detection
edges = cv2.Canny(gray, 50, 500)

plt.imshow(cv2.cvtColor(edges,cv2.COLOR_BGR2RGB))
plt.title("edge detection image")
```

```python
# Resize the image

#new dimensions

new_width = 200

new_height = 500

resized_image = cv2.resize(image,(new_width,new_height))

plt.imshow(cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB))

plt.title("Resize image")
```

```python
import cv2
import numpy as np

# Load the images
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

# Check if the images have been loaded successfully
if image1 is None or image2 is None:
    print('Failed to load the images.')
    exit()

# Ensure that both images have the same size
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Perform addition of the images
addition = cv2.add(image1, image2)

# Perform subtraction of the images
subtraction = cv2.subtract(image1, image2)

plt.imshow(cv2.cvtColor(addition,cv2.COLOR_BGR2RGB))

plt.title("Addition image")

```

```python
plt.imshow(cv2.cvtColor(subtraction,cv2.COLOR_BGR2RGB))

plt.title("Subtraction image")
```
